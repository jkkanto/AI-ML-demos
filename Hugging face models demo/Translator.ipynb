```python
from transformers import pipeline

#Lets use Finnish-English translator model from Hugging face
#pipeline takes care of everything: tokenizing data, predicting, and post-processing

translator_en_fi = pipeline("translation", model="Helsinki-NLP/opus-mt-en-fi")
translator_fi_en = pipeline("translation", model="Helsinki-NLP/opus-mt-fi-en")

```

```python
translator_en_fi.predict("All the work and no play makes Jack a dull boy.")
```

```python
translator_en_fi.predict("Life is like a box of chocolates.")
```

```python
translator_fi_en.predict("Taivas on sininen ja valkoinen ja tähtösiä täynnä.")
```

```python
translator_fi_en.predict("Joka kuuseen kurkottaa, se katajaan kapsahtaa.")
```
